## Birthday paradox
$B_n$ – moment pierwszej kolizji;
$B_n$ = k, jeśli k-ta z wrzucanych kul jest pierwszą, która
trafiła do niepustej urny

```python
x = data1.n
y = data1.B
bn = []

cnt = 0
for n in range(1000, 101000, 1000):
    sum = 0
    for k in range(50):
        sum += y[cnt*50 + k]
    bn.append(sum/50)
    cnt += 1

plt.scatter(x, y, color = "royalblue", s=1, label = "wyniki poszczególnych powtórzeń eksperymentu")
plt.scatter(range(1000, 101000, 1000), bn, color = "red", s=5, label = "średnia wartość $B_n$ dla danego n")
plt.legend(loc="upper left")
plt.show()
```

Rozbiezność poszczególnych wartości zwiększa się wraz ze wzrostem n. Koncentracja poszczególnych wyników wokół wartości średniej dla danego n zmniejsza się wraz ze wzrostem n, a rozbieznosc wartosci średnich zmniejsza się. <p>
Hipoteza: $B_n$ zbiega do wartości stałej.

## Liczba pustych urn po wrzuceniu n kul
$U_n$

```python
x = data1.n
y = data1.U
un = []

cnt = 0
for n in range(1000, 101000, 1000):
    sum = 0
    for k in range(50):
        sum += y[cnt*50 + k]
    un.append(sum/50)
    cnt += 1

plt.scatter(x, y, color = "royalblue", s=2, label = "wyniki poszczególnych powtórzeń eksperymentu")
plt.scatter(range(1000, 101000, 1000), un, color = "red", s=2, label = "średnia wartość $U_n$ dla danego n")
plt.legend(loc="lower right")
plt.show()
```

Wartości uzyskane w poszczególnych powtórzeniach są skoncentrowane wokół wartości średniej. Wartość $U_n$ rośnie wraz ze wzrostem n. <p>
Hipoteza: $U_n$ nie zbiega do wartości stałej. Średnie wyniki eksperymentu zbiegają asymptotycznie do funkcji liniowej.

## Maximum load
$L_n$ - maksymalna liczba kul w urnie po wrzuceniu n kul.

```python
x = data1.n
y = data1.L
ln = []

cnt = 0
for n in range(1000, 101000, 1000):
    sum = 0
    for k in range(50):
        sum += y[cnt*50 + k]
    ln.append(sum/50)
    cnt += 1
    
plt.scatter(x, y, color = "royalblue", s=1, label = "wyniki poszczególnych powtórzeń eksperymentu")
plt.scatter(range(1000, 101000, 1000), ln, color = "red", s=5, label = "średnia wartość $L_n$ dla danego n")
plt.legend(loc="lower right")
plt.show()
```

Rozbiezność poszczególnych wyników dla danego n utrzymuje się na zblizonym poziomie. 
Dla coraz większych n róznice w wartościach średnich nie zmieniają się znacznie. Od n = 60000 róznice w wartościach średnich dla danego n są nieznaczne. <p>
Hipoteza: $L_n$ zbiega do wartości stałej.

## Coupon collector’s problem
$C_n$ - minimalna liczba rzutów, po której w kazdej z urn jest co najmniej jedna kula.

```python
x = data2.n
y = data2.C
cn = []

cnt = 0
for n in range(1000, 101000, 1000):
    sum = 0
    for k in range(50):
        sum += y[cnt*50 + k]
    cn.append(sum/50)
    cnt += 1

plt.scatter(x, y, color = "royalblue", s=1, label = "wyniki dla poszczególnych powtórzeń eksperymentu")
plt.scatter(range(1000, 101000, 1000), cn, color = "red", s=5, label = "średnia wartość $C_n$ dla danego n")
plt.legend(loc="upper left")
plt.show()
```

Rozbiezność poszczególnych wartości zwiększa się wraz ze wzrostem n. Koncentracja poszczególnych wyników wokół wartości średniej dla danego n jest większa dla mniejszych n i zmniejsza się wraz ze wzrostem n. <p>
Hipoteza: $C_n$ nie zbiega do wartości stałej. Średnie wyniki eksperymentu zbiegają asymptotycznie do funkcji liniowej.

## Coupon collector’s brother
$D_n$ - minimalna liczba rzutów, po której w kazdej z urn są co najmniej dwie kule.

```python
x = data2.n
y = data2.D
dn = []

cnt = 0
for n in range(1000, 101000, 1000):
    sum = 0
    for k in range(50):
        sum += y[cnt*50 + k]
    dn.append(sum/50)
    cnt += 1

plt.scatter(x, y, color = "royalblue", s=1, label = "wyniki dla poszczególnych powtórzeń eksperymentu")
plt.scatter(range(1000, 101000, 1000), dn, color = "red", s=5, label = "średnia wartość $D_n$ dla danego n")
plt.legend(loc="upper left")
plt.show()
```

Rozbiezność poszczególnych wartości zwiększa się wraz ze wzrostem n. Koncentracja poszczególnych wyników wokół wartości średniej dla danego n jest większa dla mniejszych n. <p>
Hipoteza: $D_n$ nie zbiega do wartości stałej. Średnie wyniki eksperymentu zbiegają asymptotycznie do funkcji liniowej.

```python
plt.scatter(range(1000, 101000, 1000), cn, color = "lightblue", s=5, label = "średnia wartość $C_n$ dla danego n")
plt.scatter(range(1000, 101000, 1000), dn, color = "darkblue", s=5, label = "średnia wartość $D_n$ dla danego n")
plt.legend()
plt.show()
```

### $D_n - C_n$
Liczba rzutów od momentu $C_n$ potrzebna do tego, aby w kazdej urnie były co najmniej dwie kule.

```python
x = data2.n
y = data2.DminC
dmincn = []

cnt = 0
for n in range(1000, 101000, 1000):
    sum = 0
    for k in range(50):
        sum += y[cnt*50 + k]
    dmincn.append(sum/50)
    cnt += 1

plt.scatter(x, y, color = "royalblue", s=1, label = "wyniki dla poszczególnych powtórzeń eksperymentu")
plt.scatter(range(1000, 101000, 1000), dmincn, color = "red", s=5, label = "średnia wartość $D_n - C_n$ dla danego n")
plt.legend(loc="upper left")
plt.show()
```

Wraz ze wzrostem n zwiększa się rozbiezność poszczególnych wartości $D_n - C_n$ oraz wartości średnich. <p>
Hipoteza: $D_n - C_n$ nie zbiega do wartości stałej. Średnie wyniki eksperymentu zbiegają asymptotycznie do funkcji liniowej.

## Uzasadnienie nazw
####  Birthday Paradox
Mozna łatwo rozwazyć ten paradoks zastanawiając się nad prawdopodobieństwem tego, ze w pokoju, w którym znajduje się $n$ osób co najmniej dwie z nich mają urodziny tego samego dnia. Okazuje się, ze dla $n=23$ wynosi ono 0.5. Lub w drugą stronę: ile osób wystarczy rozwazyć, aby szansa, ze co najmniej dwie dzielą urodziny była równa co najmniej 0.5. Odpowiedź to 23, co wydaje się zaskakująco małą liczbą i jest sprzeczne z intuicją. 
#### Coupon collector’s problem
Rozwazmy konkurs, w którym gracz otrzymuje wygraną po zebraniu wszystkich kuponów z określonej puli. Problem polega na rozstrzygnięciu ile średnio musi kupić kuponów, by ten cel osiągnąć?


## Znaczenie Birthday Paradox w kontekście funkcji hashujących.
Standardową własnością funkcji hashujących jest bezkolizyjność - pomimo tego, ze funkcja nie jest 1-1, to prawdopodobieństwo tego, ze dwa rozne argumenty generują ten sam hash jest pomijalnie małe. Z uwagi na birthday paradox nawet przy duzej przeciwdziedzinie słabej funkcji hashującej prawdopodobienstwo kolizji jest zauwazalne. Bezkonfliktowości nie osiaga się przez zwiekszanie przeciwdziedziny do oporu, lecz przez odpowiednie zabiegi algorytmiczne. <p>
Ma on równiez znaczenie w kryptografii i jest podstawą działania tzw. ataku urodzinowego polegajacego na jest znalezieniu kolizji funkcji hashującej.
